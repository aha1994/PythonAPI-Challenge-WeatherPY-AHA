```python
# Import dependencies

from citipy import citipy
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import random
import time

from api_keys import api_key

# Latitude values range from -90 to 90
# Longitude values range from -180 to 180

lat_val = (-90, 90)
long_val = (-180, 180)
```

```python
# Generate 2000 random latitude, Longitude pairs

latitude_list = [random.uniform(-90, 90) for x in range(5000)]
longitude_list = [random.uniform(-90, 90) for x in range(5000)]
coordinates_list = zip(latitude_list, longitude_list)
cities_list = []

for x in coordinates_list:
    cities_list.append(citipy.nearest_city(x[0], x[1]).city_name)

# Remove duplicates with a set
cities_list = list(set(cities_list))

print(len(cities_list))
```

```python
# Perform Api calls
# response = requests.get(url + 'test').json()
# print(response['clouds']['all'])
# print(response['main']['humidity'])
# print(response['wind']['speed'])
# response

units = 'imperial'
url = "http://api.openweathermap.org/data/2.5/weather?appid={}&units={}&q=".format(api_key, units)

date_list = []
country_id = []
lat_list = []
long_list = []
temp_list = []
temp_min_list = []
temp_max_list = []
humidity_list = []
clouds_list = []
wind_speed_list = []
found_cities = []

print("Starting Weather Data Collection...")
print("___________________________________")
for city in cities_list:
    try:
        response = requests.get(url + city).json()
        date_list.append(response['dt'])
        country_id.append(response['sys']['country'])
        lat_list.append(response['coord']['lat'])
        long_list.append(response['coord']['lon'])
        temp_list.append(response['main']['temp'])
        temp_min_list.append(response['main']['temp_min'])
        temp_max_list.append(response['main']['temp_max'])
        humidity_list.append(response['main']['humidity'])
        clouds_list.append(response['clouds']['all'])
        wind_speed_list.append(response['wind']['speed'])
        found_cities.append(city)
        print('Record number: {} Found | {}'.format((len(date_list)), city))
        time.sleep(1)
    except (KeyError, NameError):
        print('Country NOT found...Skipping')
        time.sleep(1)
print("___________________________________")
print("Weather Data Collection Complete!")
```

```python
weather_DF = pd.DataFrame({
    'City': found_cities,
    'Country': country_id,
    'Latitude': lat_list,
    'Longitude': long_list,
    'Date': date_list,
    'Temperature': temp_list,
    'Min Temp': temp_min_list,
    'Max Temp': temp_max_list,
    'Humidity': humidity_list,
    'Cloudiness': clouds_list,
    'Wind Speed': wind_speed_list,
})
```

```python
weather_DF.to_csv('WeatherDataCollection.csv')
weather_DF.head()
```

```python
# Max Temp vs Latitude
plt.scatter(weather_DF['Latitude'], weather_DF['Max Temp'], edgecolors='black',facecolor='blue')
plt.xlabel('Degrees Latitude')
plt.ylabel('Max Temperature (F)')
plt.title('Max Temperature vs Latitude')
plt.grid(alpha=.5)
plt.savefig('Max_Temperature_vs_Latitude.png')
plt.show()
```

```python
# Humidity vs Latitude
plt.scatter(weather_DF['Latitude'], weather_DF['Humidity'], edgecolors='black',facecolor='red')
plt.xlabel('Degrees Latitude')
plt.ylabel('% Humidity')
plt.title('% Humidity vs Latitude')
plt.grid(alpha=.5)
plt.savefig('Humidity_vs_Latitude.png')
plt.show()
```

```python
# Cloudiness vs Latitude
plt.scatter(weather_DF['Latitude'], weather_DF['Cloudiness'], edgecolors='black',facecolor='orange')
plt.xlabel('Degrees Latitude')
plt.ylabel('% Cloud Cover')
plt.title('% Cloud Cover vs Latitude')
plt.grid(alpha=.5)
plt.savefig('%Cloud_Cover_vs_Latitude.png')
plt.show()
```

```python
# Wind Speed vs Latitude
plt.scatter(weather_DF['Latitude'], weather_DF['Wind Speed'], edgecolors='black',facecolor='grey')
plt.xlabel('Degrees Latitude')
plt.ylabel('Wind Speed (MPH)')
plt.title('Wind Speed (MPH) vs Latitude')
plt.grid(alpha=.5)
plt.savefig('Wind_Speed_vs_Latitude.png')
plt.show()
```
